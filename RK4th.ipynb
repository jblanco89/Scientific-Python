# Runge Kutta Fouth Order

This program performs Runge Kutta (RK) fourth order method for solving ordinary differential equation in Python.

In this Python program `t0` & `Ca0` represents initial condition to saponification reaction in Non-Isothermic CSTR reactor. `t` is final point of time on which value of `Ca` corresponding to `t` is to be calculated using Runge Kutta method. 
Step represents number of finite step before reaching to `t`, it meants, iteration number

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A=1125E-4
Ea = 48367
R = 8314E-3
V = 1450
va0 = 363E-1
vb0 = 428E-1
T0 = 29815E-2
T = 28815E-2
Cb = 0.001
ra = A*np.exp(-(Ea/R)*((1/T0) - (1/T)))*60


def dCadt (t,Ca):
  return (Ca0*va0-Ca*(va0+vb0)-ra*Ca*Cb*V)/V

def rKutta4(t0, Ca0, t, n):
  h = (t - t0) / n
  print('\n-------------SOLUTION-------------')
  print('------------------------------------')    
  print('t0          Ca0              Ca')
  print('------------------------------------')
  x = []
  y = []
  z = []
  for i in range(n):
    k1 = h * (dCadt(t0,Ca0))
    k2 = h * (dCadt((t0+h/2), (Ca0+k1/2)))
    k3 = h * (dCadt((t0+h/2),(Ca0+k2/2)))
    k4 = h * (dCadt((t0+h), (Ca0+k3)))
    k = (k1+ 2*k2 + 2*k3 + k4)/6

    Ca = Ca0 + k
    Ca0 = Ca
    t0 = t0 + h
    x.append(t0)
    z.append(Ca0)
    y.append(Ca)
    print('%.2f\t%.8f\t%.8f'% (t0,Ca0,Ca))



  print('\nAt t=%.2f, Ca=%.8f' %(t,Ca))
  fig, ax = plt.subplots()
  ax.plot(x, y, 'o-b')
  ax.set(xlabel='time (min)', ylabel='Ca (Molar)',
       title='CSTR Reactor Non-Isothermic')
  fig.set_size_inches(10,6)
  plt.show()
  #print(x,y,z)

#rKutta4(0,0.1,10,5)


```

```python
print('Enter initial conditions:')
t0 = float(input('t0 (min) = '))
Ca0 = float(input('Ca0 (M) = '))

print('Enter calculation point: ')
t = float(input('t (min) = '))

print('Enter number of steps:')
n = int(input('Iterations = '))

rKutta4(t0,Ca0,t,n)
```
