```python
#importando librerías
import numpy as np
import pandas as pd
%matplotlib 
import matplotlib.pyplot as plt
```

```python
x = np.random.randn(100)
y = np.random.randn(100)

s = 100*np.random.randn(100) ##tamaño de los indicadores
c = np.random.randn(100) ## color de las
## cmap es el mapa de colores
## consultar librería de matplotlib
plt.scatter(x,y,s,c, cmap=plt.cm.BrBG_r)
```
